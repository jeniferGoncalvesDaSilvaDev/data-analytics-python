```python
from sys import breakpointhook
#analise e visualizaçao de dados de vendas diarias de marmitas da empresa x
from matplotlib import pyplot as plt
import numpy as np
vendas_primeira_semana =[]
vendas_segunda_semana=[]
dias_da_primeira_semana=[]
dias_da_segunda_semana=[]
media_mensal=[]
def numero_marmitas_semana_01():
 try:

  for item in range(0,5):
    item = int(input("Digite o numero de marmitas vendidas, durante a primeira semana: "))
    vendas_primeira_semana.append(item)
  print(vendas_primeira_semana)



 except:
  print("Dado inválido")
def dias_da_semana_01():
  for dia in range(0,5):
      dia = str(input('Digite o dia da semana:'))
      if dia!= 'segunda' and dia!='terça' and dia!='quarta' and dia!='quinta' and dia!='sexta':
        print("Dia inválido!")
        break
      else:
        dias_da_primeira_semana.append(dia)
        print(dias_da_primeira_semana)


def media():
  media = np.mean(vendas_primeira_semana)
  media_mensal.append(media)
  print(media_mensal)
#chamando as funções
numero_marmitas_semana_01()
dias_da_semana_01()
media()
```

```python
dias_da_segunda_semana=[]
vendas_segunda_semana=[]
def numero_marmitas_semana_02():
 try:

  for item in range(0,3):
    item = int(input("Digite o numero de marmitas vendidas, durante a segunda semana: "))
    vendas_segunda_semana.append(item)
  print(vendas_segunda_semana)



 except:
  print("Dado inválido")

def dias_da_semana_02():
  for dia in range(0,3):
      dia = str(input('Digite o dia da semana:'))
      if dia!= 'segunda' and dia!='terça' and dia!='quarta' and dia!='quinta' and dia!='sexta':
        print("Dia inválido!")
        break
      else:
        dias_da_segunda_semana.append(dia)
        print(dias_da_segunda_semana)





#chamando as funçoes
numero_marmitas_semana_02()
dias_da_semana_02()

```

Visualização de dados em grafico


```python
figura = plt.figure(figsize=(60,10))
figura.suptitle('Análise de vendas de marmitas da empresa x')
figura.add_subplot(131)
plt.plot(dias_da_primeira_semana, vendas_primeira_semana, label= "Vendas diárias de marmitas na primeira semana")
plt.title("Vendas de marmitas na primeira semana ")
plt.legend()
plt.yticks([40,41,39])
plt.ylabel("Número de marmitas vendidas para a empresa x")
plt.xlabel("Dias da semana")
figura.add_subplot(132)
plt.plot(dias_da_segunda_semana, vendas_segunda_semana, label="Vendas diárias de marmitas na segunda semana",color="green")
plt.legend()
plt.yticks([32,38,39,40])
plt.ylabel("Número de marmitas vendidas para a empresa x")
plt.xlabel("Dias da semana")
```
